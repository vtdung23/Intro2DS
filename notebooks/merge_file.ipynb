```python
import pandas as pd

# 1. Đọc 2 file
df1 = pd.read_csv('../data/raw/movies_2000_2015.csv')
df2 = pd.read_csv('../data/raw/movies_2016_2025.csv')

# 2. Gộp 2 DataFrame lại (Nối dọc - Concatenation)
# axis=0 đảm bảo nối theo hàng (tức là file 2 nằm dưới file 1)
df_merged = pd.concat([df1, df2], ignore_index=True)

# 3. Lưu file đã gộp
df_merged.to_csv('../data/raw/raw_data.csv', index=False) # Lưu thành CSV
# df_merged.to_excel('file_da_gop.xlsx', index=False) # Lưu thành Excel
```
