# I. Giới thiệu Mô hình					
## 1. Tên mô hình					
**Random Forest Regressor**					
## 2. Lý do chọn mô hình					
Random Forest là một mô hình **Ensemble** dựa trên kỹ thuật **Bagging** (Bootstrap Aggregating) sử dụng nhiều cây quyết định.					
* **Tính chất:** **Phi tuyến**, có khả năng mô hình hóa mối quan hệ phức tạp giữa các đặc trưng và biến mục tiêu (rating).					
* **Phù hợp:** Rất phù hợp cho các bài toán **dự đoán định lượng (regression)**. Khả năng xử lý các đặc trưng không cần chuẩn hóa trước.					
* **Ưu điểm kỳ vọng:** Độ chính xác cao, ít bị overfit hơn so với một cây quyết định đơn lẻ, và khả năng ước tính tầm quan trọng của các đặc trưng (feature importance).					
## 3. Mục tiêu của Notebook					
* **Train model:** Huấn luyện mô hình Random Forest Regressor.					
* **Tune hyperparameters:** Tinh chỉnh các tham số quan trọng (n_estimators, max_depth) bằng **GridSearchCV** trên tập train.					
* **Evaluate model:** Đánh giá hiệu suất của mô hình tốt nhất trên tập valid.					
* **Test model:** Đánh giá cuối cùng trên tập test để kiểm tra khả năng khái quát hóa.					
* **Xuất metrics:** Lưu các chỉ số đánh giá (RMSE, MAE, R2) để đưa vào báo cáo chung.
